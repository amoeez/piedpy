This module will import the data from the pymatgen libraries for battery cathodes

```python
from pymatgen import MPRester
import pandas as pd
```

```python
api_key = ''
```

```python
protocol = MPRester(api_key)
```

```python
battery_ids = (protocol._make_request('/battery/all_ids'))
```

```python
battery_ids
```

```python
battery_data = pd.DataFrame()
for item in battery_ids:
    battery_data_row = pd.DataFrame(protocol._make_request('/battery/%s' % item))
    battery_data.append(battery_data_row)
```
